# ‚ùå –ë–ï–ó –∞–±—Å—Ç—Ä–∞–∫—Ç–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞

## –ë–∞–∑–æ–≤—ã–π –∫–ª–∞—Å—Å (–æ–±—ã—á–Ω—ã–π)

```python
# –ë–∞–∑–æ–≤—ã–π –∫–ª–∞—Å—Å (–æ–±—ã—á–Ω—ã–π)
class Payment:
    def pay(self, amount):
        pass

# –ù–∞—Å–ª–µ–¥–Ω–∏–∫ 1
class CardPayment(Payment):
    def pay(self, amount):
        print(f"Pay {amount} by card")
# –ù–∞—Å–ª–µ–¥–Ω–∏–∫ 2
class CashPayment(Payment):
    pass   # ‚ùå –∑–∞–±—ã–ª–∏ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å pay
```

## ‚ùå B–∞—Ä–∏–∞–Ω—Ç 1  (—Ä–µ–∞–ª—å–Ω—ã–π –±–∞–≥)

```python
# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ
payments = [
    CardPayment(),
    CashPayment()
]

for p in payments:
    p.pay(100)

'''
üò±
–î–µ–Ω—å–≥–∏ "–ø—Ä–∏–Ω—è–ª–∏", –Ω–æ —Ä–µ–∞–ª—å–Ω–æ –Ω–∏—á–µ–≥–æ –Ω–µ —Å–¥–µ–ª–∞–ª–∏.
–ù–∏ –æ—à–∏–±–∫–∏, –Ω–∏ –ø—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏—è.

üëâ –≠—Ç–æ —Ç–∏—Ö–∏–π –±–∞–≥ ‚Äî —Å–∞–º—ã–π –æ–ø–∞—Å–Ω—ã–π —Ç–∏–ø.
'''
```

## ‚ùå B–∞—Ä–∏–∞–Ω—Ç 2 –±–∞–≥ (–µ—â—ë —Ö—É–∂–µ)

```python
class CashPayment(Payment):
    def pay(self):
        print("Pay cash")

```

```python
p = CashPayment()
p.pay(100)

''' üí• Runtime error:
TypeError: pay() takes 1 positional argument but 2 were given
–û—à–∏–±–∫–∞ –≤—ã–ª–µ–∑–∞–µ—Ç —Å–ª–∏—à–∫–æ–º –ø–æ–∑–¥–Ω–æ. '''
```

## –ë–∞–∑–æ–≤—ã–π –∫–ª–∞—Å—Å (abstract)

```python
# –ë–∞–∑–æ–≤—ã–π –∫–ª–∞—Å—Å (abstract)
from abc import ABC, abstractmethod

class Payment(ABC):

    @abstractmethod
    def pay(self, amount):
        pass
```

## –ù–∞—Å–ª–µ–¥–Ω–∏–∫–∏

```python
# –ù–∞—Å–ª–µ–¥–Ω–∏–∫–∏
class CardPayment(Payment):
    def pay(self, amount):
        print(f"Pay {amount} by card")

class CashPayment(Payment):
    pass
```

## üîí –ß—Ç–æ –ø—Ä–æ–∏–∑–æ–π–¥—ë—Ç?

```python
# üîí –ß—Ç–æ –ø—Ä–æ–∏–∑–æ–π–¥—ë—Ç?
CashPayment()

'''
üí• –°–†–ê–ó–£ –æ—à–∏–±–∫–∞: TypeError: Can't instantiate abstract class CashPayment with abstract method pay
üëâ –ë–∞–≥ –ø–æ–π–º–∞–Ω –¥–æ –∑–∞–ø—É—Å–∫–∞ –ª–æ–≥–∏–∫–∏, –µ—â—ë –Ω–∞ —ç—Ç–∞–ø–µ —Ä–∞–∑—Ä–∞–±–æ—Ç–∫–∏.
'''
```

# üîí –ß—Ç–æ –ø—Ä–æ–∏–∑–æ–π–¥—ë—Ç?
