```python
# config.py
class AppConfig:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
            # Settings initialization / Einstellungen initialisieren
            cls._instance.theme = "light"
            cls._instance.language = "en"
        return cls._instance
```

```python
# main.py
# from config import AppConfig # This line is no longer needed as the class is now in the same file

# Get configuration object / Konfigurationsobjekt abrufen
config1 = AppConfig()
config2 = AppConfig()

# Check Singleton / Singleton prüfen
print(config1 is config2)  # True

# Use settings / Einstellungen verwenden
print("config1.theme: ", config1.theme)  # light
print("config2.theme: ", config2.theme)  # light

print("<< config2.theme = dark >>")
config2.theme = "dark"  # change theme via second object / Thema über zweites Objekt ändern

# Check that changes are visible via first object / Prüfen, dass Änderungen über erstes Objekt sichtbar sind
print("config1.theme: ", config1.theme)  # dark
print("config2.theme: ", config2.theme)  # dark


```
